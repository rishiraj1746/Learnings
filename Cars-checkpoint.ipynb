## Data Cleaning 

```python
df.head(5)
```

```python
df.drop(['Model','DriveTrain'], axis = 1, inplace = True)
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

## Data Transformation 

```python
df.head(2)
```

```python
df.rename(columns = {'Make' : 'Name'}, inplace = True)
```

```python
df.head(1)
```

```python
df['Horsepower'] = df['Horsepower'].astype('int64')
```

```python
df['MSRP'] = df['MSRP'].replace('[\$,]', '', regex=True).astype('float64')
df['Invoice'] = df['Invoice'].replace('[\$,]', '', regex=True).astype('float64')
```

```python
df.info()
```

## Data Filtering and Subsetting 

```python
df.head(2)
```

```python
filtered_df = df[df['Horsepower'] > 400 ]
```

```python
filtered_df
```

```python
subset = df[['Name','Origin']]
subset = subset.drop_duplicates()
```

```python
subset
```

## Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['EngineSize'].hist()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Type')
plt.title('Frequency of Different Car Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Origin')
plt.title('Frequency of Different Car Origins')
plt.xlabel('Origin')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Type', y='Horsepower')
plt.title('Horsepower Distribution by Car Type')
plt.xlabel('Type')
plt.ylabel('Horsepower')
plt.show()
```
